```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

baseline_data = np.array([63.0, 77.8, 29.5, 30.4, 1.6, 50.9, 0.4, 77.8, 29.5,
    63.0, 0.0, 100.0, 59.0, 60.2, 0.6, 0.5, 0.6, 0.0, 100.0, 59.0, 60.2, 90.7, 1.8,
    0.1, 0.0, 0.0, 1.8, 30.4, 90.7, 50.9])

model_outputs = np.array([45.3, 44.9, 45.3, 45.0, 45.0, 45.0, 45.0, 45.3, 45.1, 45.3, 45.1,
                         45.1, 45.6, 45.4, 43.4, 42.3, 46.5, 45.5, 43.5, 41.4, 44.9, 45.3, 
                          45.0, 45.0, 43.5, 44.5, 45.3, 45.4, 43.4, 42.6])

sns.set_theme(style="whitegrid")
plt.figure(figsize=(7,6))

#Plots the data points
sns.scatterplot(x=baseline_data, y=model_outputs, color="blue", edgecolor="black")

#Plots the theoretical line
low_lim = min(baseline_data.min(), model_outputs.min())-5
high_lim = max(baseline_data.max(), model_outputs.max())+5
plt.plot([low_lim, high_lim], [low_lim, high_lim], color="black",linewidth=2)

plt.xlabel("Baseline Toxicity Score", fontsize=11, fontweight='bold')
plt.ylabel("Model Predicted Toxicity Score", fontsize=11, fontweight='bold')
plt.title("Model Predictions vs. Baseline Toxicity Scores", fontsize=13, fontweight='bold')
         
plt.xlim(0, 105)
plt.ylim(0, 105)
plt.tight_layout()

plt.savefig("model_regression_plot.png", dpi=300)
plt.show()
```
